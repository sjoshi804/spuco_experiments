```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress


def plot_with_line_of_best_fit(df, x_col, y_col):
    """
    Plot a scatter plot with a line of best fit.

    Parameters:
    - df: pandas DataFrame
    - x_col: str, column name for X-axis
    - y_col: str, column name for Y-axis
    """
    # Plotting the data points
    plt.scatter(df[x_col], df[y_col], label='Data Points')

    # Performing linear regression to get the slope and intercept
    slope, intercept, r_value, p_value, std_err = linregress(df[x_col], df[y_col])

    # Creating a line of best fit using the slope and intercept
    line_of_best_fit = slope * df[x_col] + intercept

    # Plotting the line of best fit
    plt.plot(df[x_col], line_of_best_fit, color='red', label='Line of Best Fit')

    # Adding labels and legend
    plt.xlabel(f'{x_col} axis')
    plt.ylabel(f'{y_col} axis')
    plt.title(f'Scatter Plot with Line of Best Fit ({x_col} vs {y_col})')
    plt.legend()

    # Show the plot
    plt.show()
```
